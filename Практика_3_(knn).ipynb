```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x=np.array((1, 1, 1))
y=np.array((2, 2, 2))
z=np.array((3, 3, 3))
w=np.array((4, 4, 4))

fig = plt.figure()
ax=fig.add_subplot(111, projection = '3d')
ax.scatter(*x)
ax.scatter(*y)
ax.scatter(*z)
ax.scatter(*w)
plt.show()

print('x - y')
print('Расстояние евклида',np.linalg.norm(x-y))
print("Квадрат Евклидова расстоняния",np.linalg.norm(x-y) ** 2)
print("Расстояние Чебышева",np.linalg.norm(x-y, ord=np.inf))
print('Расстояние Хэмминга',np.linalg.norm(x-y, ord=1))
print('x - z')
print('Расстояние евклида',np.linalg.norm(x-z))
print("Квадрат Евклидова расстоняния",np.linalg.norm(x-z) ** 2)
print("Расстояние Чебышева",np.linalg.norm(x-z, ord=np.inf))
print('Расстояние Хэмминга',np.linalg.norm(x-z, ord=1))
print('x - w')
print('Расстояние евклида',np.linalg.norm(x-w))
print("Квадрат Евклидова расстоняния",np.linalg.norm(x-w) ** 2)
print("Расстояние Чебышева",np.linalg.norm(x-w, ord=np.inf))
print('Расстояние Хэмминга',np.linalg.norm(x-w, ord=1))
print('y - z')
print('Расстояние евклида',np.linalg.norm(y-z))
print("Квадрат Евклидова расстоняния",np.linalg.norm(y-z) ** 2)
print("Расстояние Чебышева",np.linalg.norm(y-z, ord=np.inf))
print('Расстояние Хэмминга',np.linalg.norm(y-z, ord=1))
print('y - w')
print('Расстояние евклида',np.linalg.norm(y-w))
print("Квадрат Евклидова расстоняния",np.linalg.norm(y-w) ** 2)
print("Расстояние Чебышева",np.linalg.norm(y-w, ord=np.inf))
print('Расстояние Хэмминга',np.linalg.norm(y-w, ord=1))
print('z - w')
print('Расстояние евклида',np.linalg.norm(z-w))
print("Квадрат Евклидова расстоняния",np.linalg.norm(z-w) ** 2)
print("Расстояние Чебышева",np.linalg.norm(z-w, ord=np.inf))
print('Расстояние Хэмминга',np.linalg.norm(z-w, ord=1))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

iris = sns.load_dataset('iris')
iris.head()

for g in range(0,3):
  n_neighbors=1
  if (g==1): n_neighbors=5
  if (g==2): n_neighbors=10
  X_train, X_test, y_train, y_test = train_test_split(iris.drop('species', axis=1), iris['species'], test_size=0.15)
  model = KNeighborsClassifier(n_neighbors)

  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)
  print(f'accuracy: {accuracy_score(y_test, y_pred)}, n_neighbors: {n_neighbors}')
  plt.figure(figsize=(10, 7))
  sns.scatterplot(x='petal_length', y='petal_width', data=iris, hue='species')
  plt.xlabel('petal_length')
  plt.ylabel('petal_width')
  plt.legend(loc=2)
  plt.grid()
  for i in range(len(y_test)):
    if (np.array(y_test)[i] != y_pred[i]):
      plt.scatter(X_test.iloc[i, 2], X_test.iloc[i, 3], color='red')
  plt.show()
```
